```python
import random

class HotelReservation:
    def __init__(self):
        self.booking = None
        self.room_prices = {
            "Standard": 2000,
            "Deluxe": 3500,
            "Suite": 5000
        }

    def book_room(self):
        if self.booking is not None:
            print(" You already have a booking.")
            return

        name = input("Enter your name: ")
        contact = input("Enter your contact number (10 digits): ")

        if not (contact.isdigit() and len(contact) == 10):
            print(" Invalid contact number!")
            return

        nights = int(input("Enter number of nights: "))

        print("\nRoom Types:")
        for r, p in self.room_prices.items():
            print(f"{r}: ₹{p}/night")

        room_type = input("Enter room type: ")

        if room_type not in self.room_prices:
            print(" Invalid room type!")
            return

        total_cost = self.room_prices[room_type] * nights
        print(f"\nTotal cost: ₹{total_cost}")

        confirm = input("Confirm booking? (yes/no): ").lower()
        if confirm == "yes":
            booking_id = "B" + str(random.randint(1000, 9999))
            self.booking = {
                "id": booking_id,
                "name": name,
                "contact": contact,
                "room_type": room_type,
                "nights": nights,
                "total": total_cost
            }
            print(f" Booking Confirmed! Your booking ID: {booking_id}")
        else:
            print(" Booking cancelled.")

    def view_booking(self):
        if self.booking:
            print("\n--- Booking Details ---")
            for k, v in self.booking.items():
                print(f"{k.capitalize()}: {v}")
        else:
            print("No booking found.")

    def cancel_booking(self):
        if self.booking:
            confirm = input("Are you sure you want to cancel? (yes/no): ").lower()
            if confirm == "yes":
                self.booking = None
                print(" Booking cancelled.")
            else:
                print(" Cancellation aborted.")
        else:
            print("No booking to cancel.")

    def menu(self):
        while True:
            print("\n--- Hotel Reservation Menu ---")
            print("1. Book Room")
            print("2. View Booking")
            print("3. Cancel Booking")
            print("4. Exit")
            choice = input("Enter choice: ")

            if choice == "1":
                self.book_room()
            elif choice == "2":
                self.view_booking()
            elif choice == "3":
                self.cancel_booking()
            elif choice == "4":
                print("Thank you for using Hotel Reservation System!")
                break
            else:
                print("Invalid choice!")

```

```python


hotel = HotelReservation()
hotel.menu()
```
