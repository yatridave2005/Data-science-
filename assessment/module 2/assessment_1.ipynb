```python
class ATM:
    def __init__(self):
        self.name = ""
        self.mobile = ""
        self.pin = ""
        self.balance = 0
        self.history = []

    def activate_card(self):
        self.name = input("Enter your name: ")
        self.mobile = input("Enter 10 digit mobile: ")
        if len(self.mobile) != 10 or not self.mobile.isdigit():
            print("Invalid mobile number")
            return
        pin1 = input("Create 4 digit PIN: ")
        pin2 = input("Re-enter PIN: ")
        if pin1 != pin2 or len(pin1) != 4 or not pin1.isdigit():
            print("PIN not set correctly")
            return
        self.pin = pin1
        deposit = int(input("Enter initial deposit (min 1000): "))
        if deposit < 1000:
            print("Minimum 1000 required")
            return
        self.balance = deposit
        self.history.append(f"Initial deposit: ₹{deposit}")
        print("Card Activated Successfully!")

    def check_pin(self):
        p = input("Enter PIN: ")
        if p == self.pin:
            return True
        else:
            print("Wrong PIN")
            return False

    def change_pin(self):
        if self.check_pin():
            new_pin = input("Enter new PIN: ")
            if len(new_pin) == 4 and new_pin.isdigit():
                self.pin = new_pin
                self.history.append("PIN Changed")
                print("PIN changed successfully")
            else:
                print("Invalid PIN format")

    def deposit_money(self):
        if self.check_pin():
            amt = int(input("Enter amount to deposit: "))
            self.balance += amt
            self.history.append(f"Deposit: ₹{amt}")
            print("Deposit successful")

    def withdraw_money(self):
        if self.check_pin():
            amt = int(input("Enter amount to withdraw: "))
            charge = 0
            if amt > 1000 and amt <= 20000:
                charge = 100
            elif amt > 20000 and amt <= 100000:
                charge = 1000
            elif amt > 100000:
                charge = 2000
            total = amt + charge
            if total > self.balance:
                print("Not enough balance")
            else:
                self.balance -= total
                self.history.append(f"Withdraw: ₹{amt} (Charge ₹{charge})")
                print(f"Withdraw successful. Charge ₹{charge}")

    def check_balance(self):
        if self.check_pin():
            print(f"Balance: ₹{self.balance}")
            self.history.append("Balance checked")

    def view_transactions(self):
        if self.check_pin():
            for t in self.history[-10:]:
                print(t)

    def menu(self):
        while True:
            print("\n1.Activate Card\n2.Change PIN\n3.Deposit\n4.Withdraw\n5.Check Balance\n6.View Last 10 Transactions\n7.Exit")
            ch = input("Enter choice: ")
            if ch == "1":
                self.activate_card()
            elif ch == "2":
                self.change_pin()
            elif ch == "3":
                self.deposit_money()
            elif ch == "4":
                self.withdraw_money()
            elif ch == "5":
                self.check_balance()
            elif ch == "6":
                self.view_transactions()
            elif ch == "7":
                print("Thank you for using ATM")
                break
            else:
                print("Invalid choice")



```

```python
# Run the ATM
atm = ATM()
atm.menu()
```
