```python
import numpy as np
from scipy import stats
import pandas as pd
```

```python
# 1.A bag contains 5 red balls, 4 blue balls, and 3 green balls. One ball is drawn at random.
# a. What is the probability of getting a red ball?
# b. What is the probability of getting a red or blue ball?
# c. What is the probability of not getting a green ball?

red = 5
blue = 4
green = 3
total = red + blue + green

# probability of red
p_red = red/total

# probability of red or blue
red_or_blue = red/total + blue/total

# probability of not green
p_not_green = 1-(green/total)

print(f"probability of red ball = {p_red}")
print(f"probability of red or blue ball = {red_or_blue}")
print(f"probability of not green ball = {p_not_green}")
```

```python
# 2.Given two vectors:
# A = [2, 3, 1] and B = [1, 0, 4]
# a. Find the dot product of A and B
# b. Are the vectors orthogonal?
A = [2, 3, 1]
B = [1, 0, 4]

# dot product of A and B
dot_product = np.dot(A,B)
print(f"dot product = {dot_product}")
# Are the vectors orthogonal?
if np.dot(A,B)== 0:
  print("vectors are orthogonal")
else:
  print("vectors are not orthogonal")
```

```python
# 3.You are given the following marks of 20 students:
# 69,35,26,55,28,99,69,50,31,32,23,28,77,80,75,45,69,61,27,48,97
# a. Find the mean, median, and mode
# b. Calculate standard deviation
# c. Comment on the symmetry of the data distribution.

marks = [69,35,26,55,28,99,69,50,31,32,23,28,77,80,75,45,69,61,27,48,97]

# a. Find the mean, median, and mode
mean = np.mean(marks)
median = np.median(marks)
mode = stats.mode(marks)
print(f"mean = {mean}")
print(f"median={median}")
print(f"mode={mode}")

# b. Calculate standard deviation
standard_deviation = np.std(marks)
print(f"standard deviation = {standard_deviation}")

# c. Comment on the symmetry of the data distribution.
skewness = stats.skew(marks)
print(f"skewness = {skewness}")

if skewness > 0:
    print("the data is right-skewed")
elif skewness < 0:
    print("the data is left-skewed")
else:
    print("the data is symmetric.")
```

```python
# 4. Use the “Student Attitude and Behavior” dataset and select the following columns: Height,
# Weight, 10th Marks, 12th Marks.
# a Find Mean, Median, Mode and Quartiles
# b Find Skewness, kurtosis and Interpretation Based on the Skewness, kurtosis.
df = pd.read_csv("Student Attitude and Behavior.csv")
cols = ["Height(CM)", "Weight(KG)", "10th Mark", "12th Mark"]
data = df[cols].apply(pd.to_numeric, errors='coerce').dropna()

# a.
# Mean, Median, Mode
mean_values = data.mean()
median_values = data.median()
mode_values = data.mode().iloc[0]

# Quartiles
Q1 = data.quantile(0.25)
Q2 = data.quantile(0.5)
Q3 = data.quantile(0.75)

print(" Mean:\n", mean_values)
print("\n Median:\n", median_values)
print("\n Mode:\n", mode_values)


# Skewness & Kurtosis

skewness = data.skew()
kurt = data.kurtosis()

print("\n Skewness:\n", skewness)
print("\n Kurtosis:\n", kurt)

print("\n Interpretation:")

for col in data.columns:
    s = skewness[col]
    k = kurt[col]

    # Skewness interpretation
    if s > 0:
        skew_text = "Positively Skewed"
    elif s < 0:
        skew_text = "Negatively Skewed"
    else:
        skew_text = "Symmetric"

    # Kurtosis interpretation
    if k > 3:
        kurt_text = "Leptokurtic"
    elif k < 3:
        kurt_text = "Platykurtic"
    else:
        kurt_text = "Mesokurtic"

    print(f"\n {col}:")
    print(f"   Skewness = {s:.2f} → {skew_text}")
    print(f"   Kurtosis = {k:.2f} → {kurt_text}")
```

```python
# 5. An investment analyst wants to assess the relationship between the stock prices of two
# companies to identify potential investment opportunities. Data: Let's consider the daily
# closing prices (in dollars) of Company A and Company B for a sample of 20 trading days:
# Company A: 45, 47, 48, 50, 52, 53, 55, 56, 58, 60, 62, 64, 65, 67, 69, 70, 72, 74, 76, 77
# Company B: 52, 54, 55, 57, 59, 60, 61, 62, 64, 66, 67, 69, 71, 73, 74, 76, 78, 80, 82, 83
# • Calculate the covariance and correlation between the stock prices of Company A
# and Company B. Interpret the value of the covariance and explain the nature of the
# relationship between the two stocks.

A = [45, 47, 48, 50, 52, 53, 55, 56, 58, 60, 62, 64, 65, 67, 69, 70, 72, 74, 76, 77]
B = [52, 54, 55, 57, 59, 60, 61, 62, 64, 66, 67, 69, 71, 73, 74, 76, 78, 80, 82, 83]

# Create DataFrame
df = pd.DataFrame({'Company A': A, 'Company B': B})

# Covariance
covariance = df['Company A'].cov(df['Company B'])

# Correlation
correlation = df['Company A'].corr(df['Company B'])

print("Covariance :", covariance)
print("Correlation :", correlation)
```
