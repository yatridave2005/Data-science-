1.  Context Window Size

- Add a validation function that raises a ValueError with a clear message if issue_type is not one

- Wrap the prompt construction in a try-except block that catches the ValueError and prints a
user-friendly error message without crashing the script.


```python

SYSTEM_PROMPT = """
You are a food delivery customer support agent.
You must respond in a professional, polite, and helpful tone.
Your responses must NOT exceed 80 words.
"""

ALLOWED_ISSUES = ["late delivery", "missing item", "wrong item"]

def validate_issue_type(issue_type):
    if issue_type not in ALLOWED_ISSUES:
        raise ValueError(
            f"Invalid issue_type: '{issue_type}'. "
            f"Allowed values are: {ALLOWED_ISSUES}"
        )


def build_user_prompt(customer_name, order_id, issue_type):
    validate_issue_type(issue_type)

    user_prompt = (
        f"Customer Name: {customer_name}\n"
        f"Order ID: {order_id}\n"
        f"Issue Type: {issue_type}\n"
        f"Please help resolve this issue."
    )

    return user_prompt


def run_test_case(customer_name, order_id, issue_type):
    try:
        user_prompt = build_user_prompt(customer_name, order_id, issue_type)

        print("\n" + "=" * 50)
        print("SYSTEM PROMPT:")
        print(SYSTEM_PROMPT.strip())

        print("\nUSER PROMPT:")
        print(user_prompt)
        print("=" * 50 + "\n")

    except ValueError as e:
        print("\nError occurred while building prompt:")
        print(str(e))
        print("-" * 50)


run_test_case(
    customer_name="Amit Shah",
    order_id="FD12345",
    issue_type="late delivery"
)

run_test_case(
    customer_name="Priya Patel",
    order_id="FD67890",
    issue_type="missing item"
)

run_test_case(
    customer_name="Rahul Mehta",
    order_id="FD99999",
    issue_type="payment issue"   
)
```

#### Task 3: Semantic Search Over Restaurant FAQs
Build a Python script that encodes a set of restaurant FAQ entries into embeddings using

embeddings for all FAQs using SentenceTransformer('all-MiniLM-L6-v2').
- Build a FAISS IndexFlatL2 index from the FAQ embeddings converted to a float32 NumPy array

- Write a function search_faq(query, k=2) that encodes the query, searches the index, and
returns the top-k most relevant FAQ strings along with their L2 distances.
- Test the function with at least two queries that use different words from the stored FAQs (e.g.
'How do I get money back?' vs an FAQ that says 'Refund policy') and print the results to confirm
that semantic matching works beyond exact keyword overlap.

```python
import numpy as np
import faiss
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')

faqs = [
    "Delivery usually takes 30 to 45 minutes depending on traffic.",
    "You can cancel your order within 5 minutes of placing it.",
    "Refunds are issued if the order is incorrect or not delivered.",
    "We may substitute items if something is unavailable, with customer approval.",
    "You can contact customer support via the app or helpline number.",
    "Orders once prepared cannot be cancelled after the cancellation window closes.",
    "Refund processing usually takes 5 to 7 business days."
]

faq_embeddings = model.encode(faqs)
faq_embeddings = np.array(faq_embeddings).astype('float32')

dimension = faq_embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)

index.add(faq_embeddings)

print("FAISS Index Size:", index.ntotal)
print("Expected FAQs:", len(faqs))

def search_faq(query, k=2):
    query_embedding = model.encode([query]).astype('float32')

    distances, indices = index.search(query_embedding, k)

    results = []
    for i in range(k):
        faq_text = faqs[indices[0][i]]
        distance = distances[0][i]
        results.append((faq_text, distance))

    return results

print("\n TEST CASE 1 ")
query1 = "How do I get money back?"
results1 = search_faq(query1, k=2)

for faq, dist in results1:
    print(f"\nFAQ: {faq}\nDistance: {dist}")


print("\n===== TEST CASE 2 =====")
query2 = "Can I stop my order after placing it?"
results2 = search_faq(query2, k=2)

for faq, dist in results2:
    print(f"\nFAQ: {faq}\nDistance: {dist}") 
    
```

#### Task 4: RAG-Powered Food Delivery Policy Q&A Pipeline
Build a Python script that implements a complete minimal RAG pipeline — chunking a policy
document, embedding the chunks, retrieving relevant context, and assembling a structured
final prompt — to answer user questions about food delivery policies.
- Define a multi-paragraph food delivery policy text of at least 300 words covering refund rules,
delivery windows, and order cancellation; split it into overlapping chunks of approximately 100
words with a 20-word overlap and print the total chunk count.
- Generate sentence-transformer embeddings for all chunks, store them in a FAISS IndexFlatL2
index, and implement a retrieve(query, k=3) function that returns the top-k most relevant
chunk strings.
- Write a build_rag_prompt(query, retrieved_chunks) function that constructs a prompt in the
following order: system role instruction › numbered retrieved context blocks › user question ›
explicit instruction to answer only from the provided context and to say 'I don't know' if the
answer is not present.
- Demonstrate the full end-to-end pipeline with a hardcoded question such as 'What is the
refund policy for missing items?' — print the retrieved chunks, the assembled prompt, and a
placeholder simulated answer string.

```python
import numpy as np
import faiss
from sklearn.feature_extraction.text import TfidfVectorizer
from sentence_transformers import SentenceTransformer
import re

model = SentenceTransformer("all-MiniLM-L6-v2")

policy_text = """
Food Delivery Policy Handbook

Our food delivery platform ensures timely and reliable service to all customers. Delivery times generally range between 30 to 60 minutes depending on restaurant preparation time, distance, and traffic conditions. In rare cases of high demand or extreme weather, delivery may take longer, but customers will be notified through the app.

Refunds are provided under specific conditions. If an order is missing items, customers are eligible for a partial or full refund depending on the value of the missing items. If the entire order is not delivered, a full refund is issued after verification. Refund requests must be submitted within 24 hours of delivery completion. Once approved, refunds are processed within 5 to 7 business days to the original payment method.

Order cancellations are allowed only within 5 minutes of placing the order, provided the restaurant has not started preparation. After preparation begins, cancellation is not possible, but customers may still raise a complaint for resolution. In cases of duplicate orders or system errors, cancellation may be processed even after preparation.

The platform prioritizes customer satisfaction and may offer compensation credits for repeated delays or service failures. All decisions are subject to internal verification by the support team.
"""

def chunk_text(text, chunk_size=100, overlap=20):
    words = text.split()
    chunks = []

    start = 0
    while start < len(words):
        end = start + chunk_size
        chunk = words[start:end]
        chunks.append(" ".join(chunk))
        start = end - overlap  # overlap

    return chunks


chunks = chunk_text(policy_text, 100, 20)

print("Total Chunks Created:", len(chunks))


chunk_embeddings = model.encode(chunks)
chunk_embeddings = np.array(chunk_embeddings).astype("float32")


dimension = chunk_embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(chunk_embeddings)

print("FAISS Index Size:", index.ntotal)

def retrieve(query, k=3):
    query_vec = model.encode([query]).astype("float32")

    distances, indices = index.search(query_vec, k)

    results = []
    for i in range(k):
        results.append((chunks[indices[0][i]], distances[0][i]))

    return results

def build_rag_prompt(query, retrieved_chunks):
    prompt = "SYSTEM: You are a food delivery policy assistant.\n"
    prompt += "Answer strictly using the provided context.\n"
    prompt += "If the answer is not in the context, say 'I don't know'.\n\n"

    prompt += "CONTEXT:\n"
    for i, chunk in enumerate(retrieved_chunks, 1):
        prompt += f"[{i}] {chunk}\n\n"

    prompt += f"QUESTION: {query}\n"
    prompt += "ANSWER:"

    return prompt

query = "What is the refund policy for missing items?"

retrieved = retrieve(query, k=3)

print("\n===== RETRIEVED CHUNKS =====")
for i, (text, dist) in enumerate(retrieved, 1):
    print(f"\nChunk {i} (Distance: {dist:.4f}):\n{text}")

prompt = build_rag_prompt(query, [r[0] for r in retrieved])

print("\n===== FINAL RAG PROMPT =====")
print(prompt)

print("\n===== SIMULATED ANSWER =====")
print("Customers are eligible for a partial or full refund if items are missing. Refunds are processed after verification within 5 to 7 business days.")
```

```python
import faiss
import numpy as np
from sentence_transformers import SentenceTransformer
from datetime import datetime

policy_text = """
Food Delivery Platform Policy

Our food delivery platform is committed to providing timely and reliable service to all customers.
Orders once placed may be cancelled within 2 minutes without any charge. After preparation begins,
cancellation may incur a partial fee depending on restaurant policy.

Refunds are issued when an order is delayed beyond the promised delivery time by more than 20 minutes,
or if the order is missing items. Customers must report issues within 24 hours of delivery to be eligible
for compensation. Refunds are processed within 5–7 business days.

Delivery delays may occur due to weather conditions, traffic congestion, or high demand periods.
In such cases, customers will be notified through the app in real time. Repeated delays may result in
compensation credits.

If a customer is unavailable at the delivery location, the delivery agent will wait for a maximum of 10 minutes.
After that, the order may be marked as delivered or returned to the restaurant depending on policy.

Hygiene and food safety are top priorities. Any complaint regarding food quality must include photographic
evidence when possible. Restaurants with repeated complaints may face suspension.

Customers are encouraged to rate their experience after every order. Ratings help improve service quality.
Loyal customers may receive promotional discounts and priority delivery during peak hours.

The platform reserves the right to update policies at any time without prior notice.
"""

model = SentenceTransformer("all-MiniLM-L6-v2")

chunks = [policy_text[i:i+300] for i in range(0, len(policy_text), 300)]
embeddings = model.encode(chunks)

dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(embeddings))

few_shot_examples = [
    ("My order is late by 1 hour", "Delivery Issue"),
    ("Food was cold and stale", "Quality Issue"),
    ("I was charged twice for my order", "Payment Issue"),
    ("Restaurant missed items in my order", "Missing Items")
]

log_file = "session_log.txt"
rag_count = 0
cls_count = 0


def log_event(mode, query, output):
    global rag_count, cls_count
    time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    with open(log_file, "a", encoding="utf-8") as f:
        f.write(f"\n[{time}] MODE: {mode}\nQUERY: {query}\nOUTPUT: {output}\n")

    if mode == "RAG":
        rag_count += 1
    else:
        cls_count += 1

def rag_query(question):
    if not question.strip():
        print(" Question cannot be empty.")
        return

    q_emb = model.encode([question])
    D, I = index.search(np.array(q_emb), k=3)

    retrieved = [chunks[i] for i in I[0]]

    prompt = f"""
You are a food delivery support assistant.

Use ONLY the context below to answer.

Context:
{retrieved[0]}
{retrieved[1]}
{retrieved[2]}

Question: {question}

Answer clearly and concisely.
"""

    print("\n Retrieved Context:")
    for r in retrieved:
        print("-", r)

    print("\n FINAL PROMPT SENT TO MODEL:\n")
    print(prompt)

    log_event("RAG", question, "RAG response generated")

def classify_complaint(text):
    if not text.strip():
        print(" Complaint cannot be empty.")
        return

    examples_text = "\n".join([f"Complaint: {c} => Category: {l}" for c, l in few_shot_examples])

    prompt = f"""
You are a complaint classification system.

Examples:
{examples_text}

Now classify the following complaint into one of:
- Delivery Issue
- Quality Issue
- Payment Issue
- Missing Items

Complaint: {text}

Return ONLY the category name.
"""

    print("\n FEW-SHOT PROMPT:\n")
    print(prompt)

    q_emb = model.encode([text])
    ex_emb = model.encode([e[0] for e in few_shot_examples])

    scores = np.dot(ex_emb, q_emb.T)
    best_idx = np.argmax(scores)

    predicted = few_shot_examples[best_idx][1]

    print("\nPredicted Category:", predicted)
    print("Closest Example:", few_shot_examples[best_idx])

    log_event("CLASSIFY", text, predicted)


def main():
    print("\n Food Delivery Help Desk Chatbot")
    print("-----------------------------------")

    while True:
        print("\n1. Ask Policy Question (RAG)")
        print("2. Classify Complaint")
        print("3. Exit")

        choice = input("\nEnter choice: ")

        if choice == "1":
            q = input("Enter your question: ")
            rag_query(q)

        elif choice == "2":
            c = input("Enter complaint: ")
            classify_complaint(c)

        elif choice == "3":
            print("\n SESSION SUMMARY")
            print("Policy Questions:", rag_count)
            print("Complaints Classified:", cls_count)
            print("👋 Exiting... Goodbye!")
            break

        else:
            print(" Invalid choice. Try again.")

if __name__ == "__main__":
    main()
```

```python
import faiss
import numpy as np
from sentence_transformers import SentenceTransformer

with open("policy.txt", "r", encoding="utf-8") as f:
    text = f.read()

chunks = text.split(". ")
chunks = [c.strip() for c in chunks if c.strip()]

model = SentenceTransformer("all-MiniLM-L6-v2")
embeddings = model.encode(chunks)

dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(embeddings))

while True:
    query = input("\nAsk question (type 'quit' to exit): ").strip()

    if query.lower() == "quit":
        print("Exiting program...")
        break

    if not query:
        print("Please enter a valid question.")
        continue

    query_emb = model.encode([query])
    D, I = index.search(np.array(query_emb), k=2)

    retrieved_chunks = []
    for idx in I[0]:
        if idx < len(chunks):
            retrieved_chunks.append(chunks[idx])

    context = "\n".join(retrieved_chunks)

    prompt = f"""
You are a helpful assistant.
Answer ONLY using the context below.

Context:
{context}

Question: {query}

If the answer is not in the context, say "Not found in policy."
"""

    print("\n Retrieved Chunks:\n", context)
    print("\n RAG PROMPT:\n", prompt)

    print("\n Simulated Answer: Based on policy, response generated.")
```
