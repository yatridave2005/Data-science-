```python
# Write a short Python script that generates a random dataset of 10 users,
# each with two features: number of daily app opens and average session duration (in minutes).
# Print the dataset as a list of dictionaries.

import random

users = []

for i in range(1, 11):
    user = {
        "User_ID": i,
        "Daily_App_Opens": random.randint(1, 50),
        "Avg_Session_Duration": round(random.uniform(2, 60), 2)
    }
    users.append(user)

print(users)
```
