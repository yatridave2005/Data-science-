```python
import numpy as np

def simple_generator(random_vector):
    """
    Generates a fake 28x28 grayscale image
    with pixel values between 0 and 1.
    """

    image = np.random.rand(28, 28)

    return image

# Example
random_vector = np.random.randn(100)
fake_image = simple_generator(random_vector)

print(fake_image.shape)
```

```python
import numpy as np

class Discriminator:

    def discriminate(self, image):

        mean_pixel = np.mean(image)

        # Simple rule:
        # lower mean → fake
        # higher mean → real

        if mean_pixel > 0.5:
            return 0.9    # likely real
        else:
            return 0.1    # likely fake


# Example
d = Discriminator()

score = d.discriminate(fake_image)

print("Probability:", score)
```

```python
import numpy as np

# Step 1: Generate random vector
random_vector = np.random.randn(100)

# Step 2: Generate fake image
fake_image = simple_generator(random_vector)

# Step 3: Discriminator evaluates image
d = Discriminator()

score = d.discriminate(fake_image)

# Step 4: Check result

if score < 0.5:
    print("Discriminator correctly identified FAKE image.")
else:
    print("Discriminator incorrectly classified image as REAL.")

print("Score:", score)
```

**ChatGPT Explanation of DCGAN, StyleGAN, and CycleGAN**

Prompt Used ::--

"Explain the difference between DCGAN, StyleGAN, and CycleGAN in 3-4 simple lines."

AI Response

DCGAN (Deep Convolutional GAN) uses convolutional neural networks to generate images and is often used for basic image generation tasks.

StyleGAN improves image quality by controlling different visual styles and features, making it popular for creating highly realistic images.

CycleGAN is designed for image-to-image translation, such as converting horses to zebras or summer scenes to winter scenes, without requiring paired training images.
