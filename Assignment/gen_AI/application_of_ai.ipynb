## **2. Zomato-Style Restaurant Q&A Chatbot**

```python
def restaurant_chatbot(name, cuisine):
    return f"Yes, {name} serves {cuisine} cuisine."

# Example Usage
restaurant = input("Enter restaurant name: ")
cuisine = input("Enter cuisine: ")


response = restaurant_chatbot(restaurant, cuisine)
print(response)
```

## **4. Spotify Playlist Filter**

```python
def filter_songs(song_list):
    return [song for song in song_list if song.startswith('S')]

# Example
songs = [
    "Shape of You",
    "Believer",
    "Senorita",
    "Sunflower",
    "Perfect"
]

print(filter_songs(songs))
```

## **5. Production Challenges of Generative AI Chatbots**

**1. Hallucinations**

A chatbot on BookMyShow or Myntra may provide incorrect movie timings or product information.

Solution: Retrieve information from verified databases before generating responses.

**2. High Cost and Slow Response**

Large AI models require expensive GPUs and may respond slowly during peak traffic.
Solution: Use optimized models, caching, and scalable cloud infrastructure.
