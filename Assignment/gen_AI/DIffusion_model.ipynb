```python
import cv2
import numpy as np

# Load grayscale image
image = cv2.imread("diffusion.png", cv2.IMREAD_GRAYSCALE)

# Check if image was loaded successfully
if image is None:
    raise FileNotFoundError("Error: diffusion.png not found or could not be loaded. Please ensure the image file is in the correct path.")

current_image = image.copy()

for step in range(1, 11):

    noise = np.random.normal(
        0,
        15,
        current_image.shape
    )

    noisy_image = current_image + noise

    noisy_image = np.clip(
        noisy_image,
        0,
        255
    ).astype(np.uint8)

    cv2.imwrite(
        f"noisy_step_{step}.png",
        noisy_image
    )

    current_image = noisy_image

print("10 noisy images saved successfully.")
```

```python
import cv2
import matplotlib.pyplot as plt

# Original image
original = cv2.imread(
    "diffusion.png", # Changed from "input.jpg" to "diffusion.png"
    cv2.IMREAD_GRAYSCALE
)

# Check if original image was loaded successfully
if original is None:
    raise FileNotFoundError("Error: diffusion.png not found or could not be loaded. Please ensure the image file is in the correct path.")

# Noisy image
noisy = cv2.imread(
    "noisy_step_10.png",
    cv2.IMREAD_GRAYSCALE
)

# Check if noisy image was loaded successfully
if noisy is None:
    raise FileNotFoundError("Error: noisy_step_10.png not found or could not be loaded.")

# Denoising
denoised = cv2.GaussianBlur(
    noisy,
    (5,5),
    0
)

# Display
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(original, cmap="gray")
plt.title("Original")

plt.subplot(1,3,2)
plt.imshow(noisy, cmap="gray")
plt.title("Noisy")

plt.subplot(1,3,3)
plt.imshow(denoised, cmap="gray")
plt.title("Denoised")

plt.show()
```

# **Risks and Ethical Concerns of Text-to-Image Models**

Risk 1: Deepfakes and Misinformation

Concern:
AI-generated images can create fake events or fake photos of people, leading to misinformation on social media.

Mitigation:
Developers can add AI-generated content watermarks and detection systems to identify synthetic images.

Risk 2: Copyright and Intellectual Property Issues

Concern:
Generated images may unintentionally imitate copyrighted artwork or artistic styles.

Mitigation:
Developers can filter training data carefully and provide attribution or licensing controls for generated content.
